
## tool co-occurence heatmap

```python
from collections import Counter

# Get top N most used tools
flat_tools = [tool for sublist in df["built_with"] for tool in sublist]
top_n = 40
top_tools = [tool for tool, _ in Counter(flat_tools).most_common(top_n)]

# Filter rows to include only those projects with at least one top tool
df_filtered = df[df["built_with"].apply(lambda x: any(t in top_tools for t in x))]

# Build filtered binary matrix
from sklearn.feature_extraction.text import CountVectorizer

df_filtered["joined_tools"] = df_filtered["built_with"].apply(lambda lst: " ".join(t for t in lst if t in top_tools))
vectorizer = CountVectorizer(vocabulary=top_tools, binary=True)
X = vectorizer.fit_transform(df_filtered["joined_tools"])
tool_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(tool_df.corr(), cmap="coolwarm", center=0, square=True, cbar=True)
plt.title("Tool Co-occurrence Heatmap (Top 40 Tools)")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

```

### Tool Co-occurrence Heatmap: Interpretation & Signals

This heatmap displays pairwise co-occurrence correlations between the **top 40 tools** used across student projects. A **higher correlation (deep red)** indicates that tools frequently appear together in the same project.

---

#### Signal-Generating Insights

**1. Core Tech Clusters Reveal Project Archetypes**
- `JavaScript`, `React`, `Node.js`, `Express.js`, `HTML/CSS` form a **Full-Stack Web Dev cluster**.
- `Flutter`, `Dart`, `Firebase`, `Android Studio` co-occur in **Mobile App Dev clusters**.
- `Python`, `Flask`, `PostgreSQL/MongoDB` suggest **Backend + ML-driven projects**.
- `Tailwind`, `HTML5`, `CSS3` highlight **UI/UX-centered work**.

📌 **Signal:** Cluster-based tool combinations can infer project types. For example, a spike in `Flutter + Firebase` hints at mobile-first innovation trends.

---

**2. Rare + Strong Pairings = Innovation Markers**
- Uncommon tools like `Gemini`, `PromptAPI`, or `Twilio` co-occurring with mainstream stacks may signal **emerging or creative tech integrations**.

📌 **Signal:** Watch for **bottom 10 tools** co-occurring with high-frequency ones — these often indicate **early adoption** or **non-obvious use cases**.

---

**3. Underutilized Tools as Novelty Zones**
- Sparse nodes like `Rust`, `C#`, or `GeminiNano` show low co-occurrence.

📌 **Signal:** Outlier combinations (e.g., `Rust + Firebase`) should be tracked for **non-conventional, high-novelty project builds**.

---

**4. JSON & GitHub = Infrastructure Signals**
- Weak but broad co-occurrence across tools shows usage as **utility infrastructure**.

📌 **Signal:** Use presence of `JSON`, `GitHub` as **normalization anchors**. Projects lacking them might be **incomplete or poorly documented**.

---

**5. Emerging Generative AI Integration**
- Tools like `Gemini`, `PromptAPI` are starting to co-occur with `Python`, `Firebase`, `JavaScript`.

📌 **Signal:** Projects combining `PromptAPI + Firebase + React` likely represent **GenAI-powered apps** — a strong trend indicator.

## dimensionality reduction for clustering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

tfidf = TfidfVectorizer(max_features=300)
X = tfidf.fit_transform(df["short_description"].fillna("")).toarray()

pca = PCA(n_components=2)
reduced = pca.fit_transform(X)

kmeans = KMeans(n_clusters=5, random_state=42).fit(reduced)
df["cluster"] = kmeans.labels_

plt.scatter(reduced[:, 0], reduced[:, 1], c=df["cluster"], cmap="tab10")
plt.title("Project Clusters by Short Description")
```

### Interpretation: Project Clusters by Short Description

This plot visualizes clusters of student projects based on the semantic similarity of their short descriptions. Each point represents a single project, embedded using NLP (e.g., TF-IDF) and reduced to 2D space via a technique like **UMAP** or **t-SNE**. Clustering (e.g., via KMeans) groups similar projects together based on language used in their descriptions.

#### What This Represents
- **Each dot** = a student project.
- **Color** = the cluster it belongs to (i.e., projects with similar themes).
- **Distance** = how similar the project descriptions are : closer = more similar.

#### Key Takeaways
- **Meaningful Groupings**: Projects form well-separated clusters, indicating that student project themes are distinguishable through short descriptions alone.
- **Thematic Overlap**: Slight overlaps between clusters suggest interdisciplinary work (e.g., AI applied to healthcare or finance).
- **Compact vs Dispersed**: Some clusters are tight (focused, repeated themes), while others are more spread out (diverse, exploratory work).
